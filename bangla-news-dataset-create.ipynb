```python
df1 = pd.read_csv('/kaggle/input/news-classification/bangla.csv')
df2 = pd.read_csv('/kaggle/input/news-classification/bbc.csv')
df3 = pd.read_csv('/kaggle/input/news-classification/dbcnews.csv')
df4 = pd.read_csv('/kaggle/input/news-classification/dhakapost.csv')
df5 = pd.read_csv('/kaggle/input/news-classification/prothomalo.csv')
df6 = pd.read_csv('/kaggle/input/news-classification/songbadprokash.csv')
df7 = pd.read_csv('/kaggle/input/news-classification/tv9bangla.csv')

```

```python
df1.rename(columns={'rm-subCat': 'headline'}, inplace=True)
df2.rename(columns={'focusIndicatorDisplayBlock': 'headline'}, inplace=True)
df3.rename(columns={'text-lg': 'headline'}, inplace=True)
df4.rename(columns={'regular-font-size': 'headline'}, inplace=True)
df5.rename(columns={'tilte-no-link-parent': 'headline'}, inplace=True)
df6.rename(columns={'category-single-text': 'headline'}, inplace=True)
df7.rename(columns={'h3': 'headline'}, inplace=True)
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)
```

```python
merged_df['category'] = 'international_politics'
```

```python
merged_df
```

```python
df = pd.read_csv('/kaggle/input/news-classification/news_classification1')
```

```python
merged_dff = pd.concat([merged_df,df], ignore_index=True)
```

```python
merged_dff['category'].unique()
```

```python
category_counts = merged_dff['category'].value_counts()
print(category_counts)
```
